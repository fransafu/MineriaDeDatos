```python
import pandas as pd
fileCSV = open('Teams.csv','r')
teams = pd.read_csv(fileCSV)
teams
```

```python
fileCSV2 = open('Players.csv','r')
players = pd.read_csv(fileCSV2)
players
```

```python
# 1) Qué jugador en un equipo con "ia" en el nombre, jugó menos de 200 minutos e hizo más de 100 pases? Imprima el apellido del jugador.

players[(players['team'].str.contains('ia')) & (players['minutes'] < 200) & (players['passes'] > 100)]['surname']
```

```python
# 2) ¿Cuál es el promedio de pases de los defensas y delanteros?
players[(players['position'] == "forward")]["passes"].mean()
```

```python
players[(players['position'] == "defender")]["passes"].mean()
```

```python
# 3) ¿Qué equipo tiene el mayor ratio de 'goalsFor' y 'goalsAgainst'? Imprimir sólo el nombre del equipo.
# teams[(teams.goalsFor == teams.goalsFor.max())]['team']
# teams.sort(['goalsFor','goalsAgainst'], ascending=[False,False])
teams[(teams['goalsFor'] == max(teams['goalsFor'])) | (teams['goalsAgainst'] == max(teams['goalsAgainst']))]["team"]
```

```python
# 4) ¿Cuántos jugadores que juegan en un equipo con ranking <10 jugaron más de 350 minutos?
jugadoresMerge = teams.merge(players)
jugadoresMerge[ (jugadoresMerge['ranking'] < 10) & (jugadoresMerge['minutes'] > 350) ]['surname'].count()
```
